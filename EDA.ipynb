```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.DataFrame()
for i in range(1,6):
    df_temp = pd.read_csv(f"Raw_Data/Creditcard_{i}.csv")
    df = pd.concat([df,df_temp], ignore_index=True)
df
```

```python
df.describe()
```

```python
df.nunique()
```

```python
print("No. of Fraudulent Trxn Datapoints:" , len(df[df["Class"]==1]))
print("No. of Legitimate Trxn Datapoints:" , len(df[df["Class"]==0]))
```

```python
sns.boxplot(data=df, x="Class", y="Amount")
plt.xlabel("Class", fontsize=10)
plt.ylabel("Amount", fontsize=10)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.savefig("Amount vs Class BoxPlot.png")
plt.show()
```

```python
upper_limit = 3000
df = df[df["Amount"]<3000]
df
```

```python
sns.boxplot(data=df, x="Class", y="Amount")
plt.xlabel("Class", fontsize=10)
plt.ylabel("Amount", fontsize=10)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.savefig("Trimmed Amount vs Class BoxPlot.png")
plt.show()
```

```python
corrmat = df.corr()
top_corr_features = corrmat.index
plt.figure(figsize=(58,58))
#plt.savefig('Corrplot.png',bbox_inches='tight')
sns.set(font_scale=4,font="Times New Roman")

#plot heat map
g=sns.heatmap(df[top_corr_features].corr(),cmap="RdYlGn", linewidths=0.1, annot=True, annot_kws={"size":35})
g.set_xticklabels(g.get_xmajorticklabels(), fontsize = 35)
g.set_yticklabels(g.get_xmajorticklabels(), fontsize = 35)
plt.savefig("Overall Correlation Heatmap.png")
```

```python
delete_columns = []
for col in df.columns:
    corr = df[col].corr(df["Class"])
    if abs(corr)>0.09:
        pass
    else:
        delete_columns.append(col)

delete_columns
```

```python
df = df.drop(delete_columns, axis=1)
df
```

```python
corrmat = df.corr()
top_corr_features = corrmat.index
plt.figure(figsize=(58,58))
#plt.savefig('Corrplot.png',bbox_inches='tight')
sns.set(font_scale=4,font="Times New Roman")

#plot heat map
g=sns.heatmap(df[top_corr_features].corr(),cmap="RdYlGn", linewidths=0.1, annot=True, annot_kws={"size":35})
g.set_xticklabels(g.get_xmajorticklabels(), fontsize = 35)
g.set_yticklabels(g.get_xmajorticklabels(), fontsize = 35)
plt.savefig("SelectedFeatureHeatmap.png")
```

```python
dfs = np.array_split(df, 5)
i=1
for chunk in dfs:
    df_temp = pd.DataFrame(chunk)
    df_temp.to_csv(f'Processed_Data/Credit_processed_{i}.csv', index=False)
    i+=1
```
